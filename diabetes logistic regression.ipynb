```python
for i in dia.columns[1:6]:
    dia[i]=dia[i].replace(0,np.median(dia[i]))
dia
```

```python
# step 5 Scaling the data
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
d1=['Outcome']
data1=sc.fit_transform(dia.drop(d1,axis=1))
```

```python
data1
```

```python
col1=dia.columns[:-1]
col1
```

```python
x=pd.DataFrame(data1,columns=col1)
x
```

```python
# model creation
```

```python
## step 1 creating independent and dependent variable
y=dia.Outcome
```

```python
## step 2 creating training and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75,random_state=42)
```

```python
## model creation
from sklearn.linear_model import LogisticRegression
clf=LogisticRegression()
clf.fit(x_train,y_train)   ## training
```

```python
## step 4 prediction
y_predict=clf.predict(x_test)
```

```python
from sklearn.metrics import *
accuracy_score(y_test,y_predict)
```

```python
sns.heatmap(confusion_matrix(y_test,y_predict),annot=True,fmt='g')
```

```python
print(classification_report(y_test,y_predict))
```

```python
y_test.value_counts()
```

```python
import matplotlib.pyplot as plt
fpr,tpr,thresholds = roc_curve(y_test,y_predict)
plt.plot(fpr,tpr,label='ROC curve')
for i, threshold in enumerate(thresholds):
    plt.annotate(f'{threshold:.2f}',(fpr[i],tpr[i]),textcoords='offset points',xytext=(0,-10),ha='center')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve with THresholds')
plt.legend(loc='lower right')
```

```python
print(thresholds)
```

```python
auc_score = roc_auc_score(y_test,y_predict)
print(f'AUC Score: {auc_score}')
```

```python
plt.figure()
plt.plot(fpr,tpr,color='blue',label=f'ROC curve (AUC = {auc_score:.2f})')
plt.plot([0,1],[0,1],color='gray',linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
custom_threshold = 1.2
y_predict_custom_threshold = (y_predict >= custom_threshold).astype(int)
accuracy =accuracy_score(y_test,y_predict_custom_threshold)
print(f'Accuracy at threshold{custom_threshold}:{accuracy}')
```
